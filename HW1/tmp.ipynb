```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('data.xlsx',sheet_name='Sheet2')
```

```python
fig = plt.figure()
plt.scatter(df['size_of_array'],df['insert_sort_time'],marker='x',label='insertion sort')
plt.scatter(df['size_of_array'],df['merge_sort_time'],marker='o',label='merge sort');
plt.xlabel('size_of_array')
plt.ylabel('time in nanoseconds')
plt.legend()
plt.savefig('output.png')
```
